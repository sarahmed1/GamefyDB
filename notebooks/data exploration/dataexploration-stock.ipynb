```python
# Remove empty rows and "Unnamed" columns only
df = df.dropna(how='all')
df = df.loc[:, ~df.columns.str.contains(r'^Unnamed', na=False)].copy()
```

```python
# Clean numeric money columns safely
money_cols = ['Unit Price', 'Total Amount']

for col in money_cols:
    if col in df.columns:
        # keep original text version (no data loss)
        df[f'{col}_raw'] = df[col]

        s = df[col].astype(str).str.strip()

        # remove repeated header text inside data rows
        s = s.mask(s.str.fullmatch(col, case=False, na=False))

        # clean currency/text noise
        s = (
            s
            .str.replace(r'\s*TND', '', regex=True)
            .str.replace(',', '.', regex=False)
            .str.replace(r'[^0-9.\-]', '', regex=True)
        )

        # convert to numeric (invalid stays NaN, no crash)
        df[col] = pd.to_numeric(s, errors='coerce')

# Optional: remove rows that are clearly repeated headers in data
if set(money_cols).issubset(df.columns):
    repeated_header = (
        df['Unit Price_raw'].astype(str).str.contains('Unit Price', case=False, na=False) &
        df['Total Amount_raw'].astype(str).str.contains('Total Amount', case=False, na=False)
    )
    df = df.loc[~repeated_header].reset_index(drop=True)
```

```python
# Keep exactly needed columns (edit if you want to keep/remove raw columns)
needed_cols = [
    'Cashier', 'Date', 'Item Name', 'Category', 'In/Out',
    'Quantity', 'Unit Price', 'Total Amount'
]
needed_cols = [c for c in needed_cols if c in df.columns]
df_final = df[needed_cols].copy()
```

```python
# Clean types
df_final['Date'] = pd.to_datetime(df_final['Date'], errors='coerce')
df_final['Quantity'] = pd.to_numeric(df_final['Quantity'], errors='coerce')
df_final['Unit Price'] = pd.to_numeric(df_final['Unit Price'], errors='coerce')
df_final['Total Amount'] = pd.to_numeric(df_final['Total Amount'], errors='coerce')
```

```python
# Remove rows missing core fields
df_final = df_final.dropna(subset=['Date', 'Item Name', 'Quantity', 'Unit Price', 'Total Amount'])

# Optional: remove zero/negative quantity rows
df_final = df_final[df_final['Quantity'] > 0]
```

```python
# Preview clean table
display(df_final.head(20))
print(df_final.info())
```

```python
# 1) Daily trend
daily = df_final.groupby(df_final['Date'].dt.date)['Total Amount'].sum()
plt.figure(figsize=(8,4))
daily.plot()
plt.title('Daily Total Amount')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.tight_layout()
plt.show()
```

```python
# 2) By category
cat = (df_final.groupby('Category', as_index=False)['Total Amount']
       .sum()
       .sort_values('Total Amount', ascending=True))

plt.figure(figsize=(8,4))
plt.scatter(cat['Total Amount'], cat['Category'], s=80)
plt.title('Total Amount by Category (Dot Plot)')
plt.xlabel('Total Amount')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# 4) In/Out comparison
io = df_final.groupby('In/Out', as_index=False)['Total Amount'].sum()

plt.figure(figsize=(5,5))
plt.pie(
    io['Total Amount'],
    labels=io['In/Out'],
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.45}  # donut
)
plt.title('In vs Out Share of Total Amount')
plt.tight_layout()
plt.show()
```
